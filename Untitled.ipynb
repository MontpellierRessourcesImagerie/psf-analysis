```python
path = '/home/shaswati/Documents/PSF/YC-63x-1v4-880-banana/plots/radial_profiles_63x_1.json'


import json
import matplotlib.pyplot as plt
import numpy as np

# Replace 'path/to/your/file.json' with the path of your JSON file
with open(path, 'r') as f:
    data = json.load(f)

# `data` is a dictionary with integers as keys and lists of integers as values

# Preparing data for the graph
colors = plt.cm.jet(np.linspace(0, 1, len(data)))  # Create a color chart

#Creating the graph
for (key, y_values), color in zip(data.items(), colors):
    x_values = range(len(y_values)) # The abscissa are simply the rank of each ordinate 
    plt.plot(x_values, y_values, color=color)

plt.xlabel('Ordinate Index')
plt.ylabel('Value')
plt.title('Graph of ordered lists')
plt.legend()
plt.show()
```

```python
import json
import os
import matplotlib.pyplot as plt
import numpy as np

# Load the categorized angles from the JSON file
json_file_path = "/home/shaswati/Documents/PSF/63x-confocal-ok/categorized_angles_02.json"  # Replace with the actual path
with open(json_file_path, 'r') as json_file:
    categorized_angles_list = json.load(json_file)

print(categorized_angles_list) 

# Extract categories and counts
categories = categorized_angles_list[0].keys()
counts = [[category_value for category_value in category.values()] for category in categorized_angles_list]


# Plot the bar chart
fig, ax = plt.subplots()
bar_width = 0.2
opacity = 0.8

for i, category in enumerate(categories):
    x = np.arange(len(counts[i])) + i * bar_width
    ax.bar(x, counts[i], bar_width, label=category)

ax.set_xlabel('Images')
ax.set_ylabel('Counts')
ax.set_title('Categorized Angles')
ax.set_xticks(np.arange(len(counts[0])) + 0.5 * bar_width)
ax.set_xticklabels([str(j + 1) for j in range(len(counts[0]))])
ax.legend()

plt.show()
```
